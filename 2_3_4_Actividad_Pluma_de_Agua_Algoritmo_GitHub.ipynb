## 2. Situación problema

Un hogar recibe una cierta cantidad de **plumas de agua**.  
Se sabe que:

- **1 pluma = 0.025 litros por segundo**

Con esta información, el estudiante debe construir un programa que permita calcular:

1. el caudal total en litros por segundo;
2. el consumo diario en litros;
3. el consumo mensual en litros (suponiendo 30 días);
4. una clasificación del consumo:
   - **Bajo**: menos de 1000 litros por día
   - **Medio**: entre 1000 y 3000 litros por día
   - **Alto**: más de 3000 litros por día

---

```python
# ==========================================================
# ACTIVIDAD: MODELAR LA "PLUMA DE AGUA" CON PYTHON
# ==========================================================

# 1. Solicitar al usuario el número de plumas
plumas = int(input("Ingrese el número de plumas de agua asignadas al hogar: "))

# 2. Definir constantes del problema
caudal_pluma = 0.025      # litros por segundo que representa una pluma
segundos_dia = 86400      # segundos que tiene un día (60 * 60 * 24)
dias_mes = 30             # días supuestos por mes

# 3. Calcular el caudal total
caudal_total = plumas * caudal_pluma  # litros por segundo

# 4. Calcular el consumo diario
consumo_diario = caudal_total * segundos_dia  # litros por día

# 5. Calcular el consumo mensual
consumo_mensual = consumo_diario * dias_mes  # litros por mes

# 6. Clasificar el consumo
if consumo_diario < 1000:
    nivel_consumo = "Bajo"
elif 1000 <= consumo_diario <= 3000:
    nivel_consumo = "Medio"
else:
    nivel_consumo = "Alto"

# 7. Mostrar resultados
print("\n===== RESULTADOS =====")
print(f"Número de plumas:       {plumas}")
print(f"Caudal total:           {caudal_total:.4f} litros/segundo")
print(f"Consumo diario:         {consumo_diario:.2f} litros/día")
print(f"Consumo mensual:        {consumo_mensual:.2f} litros/mes")
print(f"Clasificación:          {nivel_consumo}")
```

#Verificación Conceptual
1.**¿Qué representa físicamente la variable principal del problema?**

La variable plumas representa la cantidad de conductos de agua asignados a un hogar. Cada "pluma" es una medida colonial del caudal equivalente a 0.025 litros por segundo; la variable principal del cálculo es caudal_total, que expresa cuánta agua fluye por segundo hacia ese hogar.

2. **¿Por qué este ejercicio conecta historia, cultura e ingeniería?**

Porque la palabra "pluma" proviene del sistema de distribución de agua colonial, donde se usaban plumas de ave para medir el flujo. Ese término cotidiano del Caribe colombiano tiene raíces hidráulicas históricas que hoy podemos modelar computacionalmente, mostrando que la programación también nace del contexto cultural y local.

3. **¿Qué operaciones aritméticas fueron necesarias?**
Multiplicación en todos los pasos:

plumas × caudal_pluma → caudal total

caudal_total × segundos_dia → consumo diario

consumo_diario × dias_mes → consumo mensual

4. **¿Qué estructura condicional utilizó y por qué?**

Se usó if / elif / else porque hay tres categorías mutuamente excluyentes (Bajo, Medio, Alto) con rangos definidos. Esta estructura evalúa las condiciones en orden y ejecuta solo el bloque que corresponde al valor de consumo_diario.

5. **¿Cómo interpretaría el resultado en un contexto real?**

Si un hogar tiene, por ejemplo, 2 plumas: su caudal es 0.05 L/s, consume 4.320 litros al día y 129.600 litros al mes, clasificado como "Alto". Esto permitiría a una empresa de acueducto estimar facturación, detectar derroche o planear la red de distribución.

```python
#Extensión
# Consumo semanal
consumo_semanal = consumo_diario * 7
print(f"Consumo semanal:        {consumo_semanal:.2f} litros/semana")

# Conversión a metros cúbicos
consumo_diario_m3 = consumo_diario / 1000
consumo_mensual_m3 = consumo_mensual / 1000
print(f"Consumo diario (m³):    {consumo_diario_m3:.3f} m³/día")
print(f"Consumo mensual (m³):   {consumo_mensual_m3:.3f} m³/mes")

# Estimación de costo (tarifa ejemplo: $4.50 por m³)
tarifa = float(input("Ingrese la tarifa por metro cúbico ($): "))
costo_mensual = consumo_mensual_m3 * tarifa
print(f"Costo mensual estimado: ${costo_mensual:.2f}")
```
